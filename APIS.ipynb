```python
response2 = requests.get('https://api.github.com/events')
```

```python
data2 = pd.DataFrame(response2.json())
```

```python
data2.head()
```

En el caso de arriba tenemos nested contents, lo cual es un problema a la hora de crear el dataframe. Para evitar esto utilizamos una función de pandas:

```python
from pandas.io.json import json_normalize
```

Lo metemos en una variable

```python
results2 = response2.json()
```

Hacemos flattening de los datos:

```python
flattened_data = json_normalize(results2)
```

Nos hace automáticamente el dataframe:

```python
flattened_data.head()
```
