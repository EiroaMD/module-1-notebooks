## Buscar columnas donde el 90% de los valores son iguales:

Si el 90% de los valores son los mismos. Es decir, si el mínimo de mi columna es igual al percentil 90 ==> lo consigo

Hacer un loop para ello:

```python
low_variance = []
for col in data._get_numeric_data():
    minimum = min(data[col])
    ninety_perc = np.percentile(data[col],90)
    if minimum == ninety_perc:
        low_variance.append(col)
print(low_variance)
```

## Buscar outliers:

Usamos describe, que te da los estadísticos para todas las columnas.

```python
stats = data.describe()
stats
```

Para leerlo más bonito, hmacemos transpose

```python
stats = data.describe().T
stats.head()
```

Añadir rango intercuartílico:

```python
stats['IQR'] = stats['75%'] - stats['25%']
stats.head()
```

```python
stats['low'] = stats['25%'] - 1.5*stats['IQR']
stats['high'] = stats['75%'] + 1.5*stats['IQR']
```

```python
outliers = pd.DataFrame(columns=data.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = data[(data[col] < lower) |
                  (data[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
    
```
