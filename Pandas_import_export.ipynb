### Exportar un df a una base de datos

```python
pip install pandas-gbq
```

Primero creo el df aquí for explanation purposes:

```python
query='''
SELECT
end_station_name,
end_station_longitude,
end_station_latitude,
COUNT(*) AS total_trips
FROM
   `bigquery-public-data.new_york.citibike_trips`
GROUP BY
    end_station_name,
    end_station_longitude,
    end_station_latitude
ORDER BY
   total_trips DESC
'''
```

```python
query_job = client.query(query=query)
df = query_job.to_dataframe()
df.head()
```

Ahora sí que vamos a a exportar

```python
full_table_id = 'car_dealership.new_table_DE'
project_id = 'ironhack-data-analytics'

df.to_gbq(full_table_id,project_id=project_id)
```
